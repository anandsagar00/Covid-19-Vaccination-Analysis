```python
figure(figsize=(2560/200,1600/200), dpi=200)
plt.pie(freq.values(),labels=freq.keys(),rotatelabels=True)
plt.show()
```

```python
s=0 
for a,b in freq.items():
    s+=b
for a,b in freq.items():
    freq[a]=(b/s)*100
freq1={}
for a,b in freq.items():
    k=a+" "+str(b)[:5]+" %"
    freq1[k]=b
```

```python
fig, ax = plt.subplots(figsize=(18, 12), subplot_kw=dict(aspect="equal"))
wedges, texts = ax.pie(freq1.values(), wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(list(freq1.keys())[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw,fontsize=16)

plt.savefig('shareofvaccines.png')
# plt.savefig('shareofvaccines.pdf')
plt.show()

'''
Question ? 
Why pfizer is dominating over rest of the vaccines ? 
'''
```

```python
def visualise(country,age_group):
    figure(figsize=(2560/200,1600/200), dpi=200)
    dff=df2[(df2['age_group']==age_group) & (df2['location']==country)]
    dff.head()
    sns.lineplot(x='date',y='people_vaccinated_per_hundred',data=dff,label='vaccinated_per_hundred')
    sns.lineplot(x='date',y='people_fully_vaccinated_per_hundred',data=dff,label='people_fully_vaccinated_per_hundred')
    sns.lineplot(x='date',y='people_with_booster_per_hundred',data=dff,label='people_with_booster_per_hundred')
    list_date=list(dff.date.unique())
    plt.legend(loc=0)
    plt.xticks(list_date[0:len(list_date):50],rotation=45)
    plt.title("Analysis of age "+age_group+" of "+country)
    plt.savefig(country+" "+age_group+".png")
    
```

```python
lst_dt=list(df_uk.date.unique())
```

```python
df_uk.head()
```

```python
freq_uk_v={}
freq_uk_fv={}
for ldt in lst_dt:
    if ldt not in freq_uk_v.keys():
        freq_uk_v[ldt]=0
        freq_uk_fv[ldt]=0
```

```python
for i in range(0,df_uk.shape[0]):
    all_data=dict(df_uk.iloc[i])
    if (all_data['date'] in freq_uk_v.keys() and (all_data['people_vaccinated_per_hundred']!=0 or all_data['people_fully_vaccinated_per_hundred']!=0)) :
        freq_uk_v[all_data['date']]+=all_data['people_vaccinated_per_hundred']
        freq_uk_fv[all_data['date']]+=all_data['people_fully_vaccinated_per_hundred']
```

```python
lst_dt=list(df_isr.date.unique())
freq_isr_v={}
freq_isr_fv={}
for ldt in lst_dt:
    if ldt not in freq_isr_v.keys():
        freq_isr_v[ldt]=0
        freq_isr_fv[ldt]=0

for i in range(0,df_isr.shape[0]):
    all_data=dict(df_isr.iloc[i])
    if (all_data['date'] in freq_isr_v.keys() and (all_data['people_vaccinated_per_hundred']!=0 or all_data['people_fully_vaccinated_per_hundred']!=0)) :
        freq_isr_v[all_data['date']]+=all_data['people_vaccinated_per_hundred']
        freq_isr_fv[all_data['date']]+=all_data['people_fully_vaccinated_per_hundred']
```
