```python
#this function recieves a df and returns the 
def outlier_range(x):
    q1=np.percentile(x,25)
    q3=np.percentile(x,75)
    iqr=q3-q1
    lower=q1-(1.5*iqr)
    upper=q3+(1.5*iqr)
    return (lower,upper)
```

```python
def extreme_outlier_range(x):
    q1=np.percentile(x,25)
    q3=np.percentile(x,75)
    iqr=q3-q1
    lower=q1-(3*iqr)
    upper=q3+(3*iqr)
    return (lower,upper)
```

```python
def remove_outliers(df, column, min_val, max_val):
    col_values = df[column].values
    df[column] = np.where(((df[column].isnull())|(df[column]<=min_val)|(df[column]>=max_val)),max_val, col_values)
    return df
```

```python
def increase_by_one(df,column):
    col_values = df[column].values
    df[column] = np.where((df[column].isnull()),col_values, col_values+1)
    return df
```

# Key Takeaways from pre-processing 'new_cases' column : 

- The data was highly skewed because of which there were too many outliers and it was becoming too difficult to handle outliers
- We tried all methods to deal with outliers such as limiting them in a boundary,but due to skewness it was getting out of hand
- Then we transformed the feature using log , which removed the skewness in our data as well as removed outliers
- The major problem in using log was to handle 0 so I increased the each value by 1
- To handle nan/null values I replaced it with the next non-null data using method="pad" in fillna function,since the data can become wrong if we replace it with any constant or mean,mediam,mode as in beginning we have very few cases and later it is increasing vigrously , so the only method to fill was using the next value as the data given is sorted a/c to date so it will be somwhow closer to actual observation

```python
#A function which returns the list of ouliers in the data
def outlier_detector(df1):
    q1=np.percentile(df1,25)
    q3=np.percentile(df1,75)
    iqr=q3-q1
    floor=q1-(1.5*iqr)
    ceil=q3+(1.5*iqr)
    outlier_index=list(df1.index[(df1<floor) | (df1>ceil)])
    outlier_values=list(df1[outlier_index])
    return outlier_values
```

```python
def clean_my_col(df,column,drop=False,save=False,zero=False,mean=False,median=False,mode=False,pad=False,bfill=False):
    
    #visualising the distribution and box plot before cleaning
#     figure(figsize=(8,6), dpi=80)
#     plt.subplot(2,1,1)
#     sns.kdeplot(df[column])
#     plt.subplot(2,1,2)
#     sns.boxplot(df[column])

    print("Skewness of data is ",df[column].skew())

    #increased all values in a column by 1, so that 0 doesn't cause error in log
    print("Increasing all values in a column by 1, so that 0 doesn't cause error in log")
    df=increase_by_one(df,column)
    print("Increasing done")
    #Taking log transform
    
    ncolumn=column+"_log"#name of new column
    
    print("Taking log transfrom")
    df[ncolumn]=np.log(df[column])
    print("Log transformation done")

    print("New Skewness of data is : ",df[ncolumn].skew())

    print("Current null values : ",df[ncolumn].isnull().sum())
    if pad==True:
        print("Filling the null values with next available values.")
        df[ncolumn].fillna(method="pad",inplace=True)
        print("Filling done.")
        print("New Null values : ",df[ncolumn].isnull().sum())
    elif zero==True:
        print("Filling the null values with 0.")
        df[ncolumn].fillna(0,inplace=True)
        print("Filling done.")
        print("New Null values : ",df[ncolumn].isnull().sum())
    elif mean==True:
        print("Filling the null values with mean.")
        df[ncolumn].fillna(df[ncolumn].mean(),inplace=True)
        print("Filling done.")
        print("New Null values : ",df[ncolumn].isnull().sum())
    elif median==True:
        print("Filling the null values with median.")
        df[ncolumn].fillna(df[ncolumn].median(),inplace=True)
        print("Filling done.")
        print("New Null values : ",df[ncolumn].isnull().sum())
    elif mode==True:
        print("Filling the null values with mode.")
        df[ncolumn].fillna(df[ncolumn].mode()[0],inplace=True)
        print("Filling done.")
        print("New Null values : ",df[ncolumn].isnull().sum())
    elif bfill==True:
        print("Filling the null values with previous value.")
        df[ncolumn].fillna(method="bfill",inplace=True)
        print("Filling done.")
        print("New Null values : ",df[ncolumn].isnull().sum())

    figure(figsize=(2560/200,1600/200), dpi=200)
    plt.subplot(2,2,1)
    sns.kdeplot(df[column])
    plt.subplot(2,2,2)
    sns.boxplot(df[column])
    plt.subplot(2,2,3)
    sns.kdeplot(df[ncolumn])
    plt.subplot(2,2,4)
    sns.boxplot(df[ncolumn])
    if save==True:
        plt.savefig(ncolumn+".png")
        plt.savefig(ncolumn+".pdf")

    #Now we don't require the old col so we drop it
    if drop==True:
        df.drop(columns=[column],axis=1,inplace=True)
    return df
```
