```python
import numpy as np,pandas as pd,matplotlib.pyplot as plt,seaborn as sns
```

```python
from datetime import datetime
```

```python
custom_date_parser = lambda x: datetime.strptime(x, "%Y-%m-%d")
```

```python
df=pd.read_csv('review_3_data.csv',parse_dates=['date'],date_parser=custom_date_parser)
```

```python
df_india=df[(df.location=='India')]
```

```python
df_india_2020=df_india[pd.DatetimeIndex(df_india['date']).year==2020]
```

```python
df_india_2021=df_india[pd.DatetimeIndex(df_india['date']).year==2021]
```

# GDP

```python
df_india.columns
```

```python
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
figure(figsize=(2560/200,1600/200), dpi=200)
plt.subplot(1,2,1)
sns.lineplot(df_india_2020.date,df_india_2020.gdp_per_capita)
plt.subplot(1,2,2)
sns.lineplot(df_india_2021.date,df_india_2021.gdp_per_capita)
```

```python
figure(figsize=(2560/200,1600/200), dpi=200)
plt.subplot(1,2,1)
sns.lineplot(df_india_2020.date,df_india_2020.total_deaths_log)
plt.subplot(1,2,2)
sns.lineplot(df_india_2021.date,df_india_2021.total_deaths_log)
```

```python
figure(figsize=(2560/200,1600/200), dpi=200)
plt.subplot(1,2,1)
sns.lineplot(df_india_2020.date,df_india_2020.total_vaccinations_in_million)
plt.subplot(1,2,2)
sns.lineplot(df_india_2021.date,df_india_2021.total_vaccinations_in_million)
```

```python
figure(figsize=(2560/200,1600/200), dpi=200)
plt.subplot(3,2,1)
sns.lineplot(df_india_2020.date,df_india_2020.gdp_per_capita)
plt.xlabel("Date of 2020")
plt.ylabel("GDP per capita")
plt.subplot(3,2,2)
sns.lineplot(df_india_2021.date,df_india_2021.gdp_per_capita)
plt.xlabel("Date of 2021")
plt.ylabel("GDP per capita")
plt.subplot(3,2,3)
sns.lineplot(df_india_2020.date,df_india_2020.total_deaths_log)
plt.xlabel("Date of 2020")
plt.ylabel("Total Deaths (in log 10)")
plt.subplot(3,2,4)
sns.lineplot(df_india_2021.date,df_india_2021.total_deaths_log)
plt.xlabel("Date of 2021")
plt.ylabel("Total Deaths (in log 10)")
plt.subplot(3,2,5)
sns.lineplot(df_india_2020.date,df_india_2020.total_vaccinations_in_million)
plt.xlabel("Date of 2020")
plt.ylabel("Total Vaccinations in Million")
plt.subplot(3,2,6)
sns.lineplot(df_india_2021.date,df_india_2021.total_vaccinations_in_million)
plt.xlabel("Date of 2021")
plt.ylabel("Total Vaccinations in Million")
plt.show()
```
